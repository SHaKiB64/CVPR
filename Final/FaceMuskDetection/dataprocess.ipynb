```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
haar_data = cv2.CascadeClassifier('data.xml')
```

```python
capture = cv2.VideoCapture(0)
withoutMask = []
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y),(x+w, y+h), (255,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(withoutMask))
            if len(withoutMask) < 200:
                withoutMask.append(face)
        cv2.imshow("Image", img)
        if cv2.waitKey(2) == 27 or len(withoutMask) >= 200:
            break
capture.release()
cv2.destoryAllWindows()
```

```python
capture = cv2.VideoCapture(0)
withMask = []
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y),(x+w, y+h), (255,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(withMask))
            if len(withMask) < 200:
                withMask.append(face)
        cv2.imshow("Image", img)
        if cv2.waitKey(2) == 27 or len(withMask) >= 200:
            break
capture.release()
cv2.destoryAllWindows()
```

```python
capture = cv2.VideoCapture(0)
incorrectMask = []
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y),(x+w, y+h), (255,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(incorrectMask))
            if len(incorrectMask) < 200:
                incorrectMask.append(face)
        cv2.imshow("Image", img)
        if cv2.waitKey(2) == 27 or len(incorrectMask) >= 200:
            break
capture.release()
cv2.destoryAllWindows()
```

```python
np.save('withoutMask.npy', withoutMask)
```

```python
np.save('withMask.npy', withMask)
```

```python
np.save('incorrectMask.npy', incorrectMask)
```
